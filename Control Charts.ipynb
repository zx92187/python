```python
TSE['Date']=pd.to_datetime(TSE['Date'])
```

```python
TSE.info()
```

```python
TSE['Close'].mean()
```

```python
TSE['Close_mean']=TSE['Close'].mean()
TSE.head()
```

```python
TSE['Close_std']=TSE['Close'].std()
TSE.head()
```

```python
TSE['Close_Upper_1std']= TSE['Close_mean'] + TSE['Close_std']
TSE['Close_Upper_2std']= TSE['Close_mean'] + TSE['Close_std']*2
TSE['Close_Upper_3std']= TSE['Close_mean'] + TSE['Close_std']*3
TSE.head()
```

```python
TSE['Close_Lower_1std']= TSE['Close_mean'] - TSE['Close_std']
TSE['Close_Lower_2std']= TSE['Close_mean'] - TSE['Close_std']*2
TSE['Close_Lower_3std']= TSE['Close_mean'] - TSE['Close_std']*3
TSE.head()
```

```python
TSE['Close'].plot(marker='o')
TSE['Close_mean'].plot(color='r')
TSE['Close_Upper_3std'].plot(color='b',ls='--')
TSE['Close_Lower_3std'].plot(color='k',ls='--')
```
